```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import plotly as py
import plotly.graph_objs as go
import plotly.express as px
import ipywidgets as widgets
```

```python
df = pd.read_csv('cut_df.csv')
df.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
df
```

```python
df['NAICSCode'].fillna(0, inplace=True)
```

```python
df['Zip'] = df['Zip'].astype(int)
df['Zip'] = df['Zip'].astype(int)
df['NAICSCode'] = df['NAICSCode'].astype(int)
df['DateApproved']= pd.to_datetime(df['DateApproved'])
```

Ranges for Salary:

    a = 5M - 10M -- 5M
    b = 2M - 4.9M -- 2M

    d = 350,000 - 999,999 -- 350k
    e = 150,000 - 349,999 -- 150k
    

```python
df.LoanRange = df['LoanRange'].str.replace(' ', '',)
```

```python
df.LoanRange = df['LoanRange'].str.replace('a', '')
df.LoanRange = df['LoanRange'].str.replace('b', '')
df.LoanRange = df['LoanRange'].str.replace('c', '')
df.LoanRange = df['LoanRange'].str.replace('d', '')
df.LoanRange = df['LoanRange'].str.replace('e', '')
```
